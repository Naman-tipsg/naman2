```python
def even_odd(number):
    if number % 2 == 0:
        return "even"
    else:
        return "odd"

while True:
    try:

        num = int(input("Enter a number  "))


        result = even_odd(num)
        print(f"The number {num} is {result}.")
    except ValueError:
        print("Exiting the program.")
        break
```

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers."
    elif n == 0 or n == 1:
        return 1
    else:
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result


while True:
    try:

        num = int(input("Enter a number to calculate its factorial (or type 'exit' to quit): "))


        result = factorial(num)
        print(f"The factorial of {num} is {result}.")
    except ValueError:
        print("Exiting the program.")
        break
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        return f"Hello, my name is {self.name} and I am {self.age} years old."

    def birthday(self):
        self.age += 1
        return f"Happy Birthday, {self.name}! Now you are {self.age} years old."


person1 = Person("Alice", 30)


print(person1.greet())
print(person1.birthday())
```
